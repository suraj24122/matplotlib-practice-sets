```python
import matplotlib.pyplot as plt
```

```python
years = list(range(2010, 2021))
consumption = [210, 220, 235, 240, 250, 260, 275, 280, 290, 300, 310]  # in gigawatt-hours
```

```python
plt.plot(years, consumption, label = "Electricity bill", color = "blue", marker = "s")
plt.title("Electricity Consumption in India (2010–2020)")
plt.xlabel("years")
plt.ylabel("consumption")
plt.grid(True)
plt.legend()
plt.show()
```
