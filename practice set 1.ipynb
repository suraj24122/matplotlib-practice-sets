```python
import matplotlib.pyplot as plt

# Data
years = list(range(2010, 2021))
population = [5.5, 5.8, 6.0, 6.3, 6.7, 7.1, 7.4, 7.8, 8.3, 8.7, 9.2]

# Plot
plt.plot(years, population, color='green', linestyle='--', marker='o', label="Population Growth")
plt.title("Year-wise Population Growth")
plt.xlabel("Years")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
