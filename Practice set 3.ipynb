```python
import matplotlib.pyplot as plt
```

```python
departments = ["HR", "Sales", "Marketing", "IT", "Finance"]
revenue = [120, 300, 250, 400, 220]
```

```python
plt.plot(departments, revenue, label = "Revenue by department", linestyle = "--", color = "blue", marker = ">")
plt.title("Quarterly Department Revenue")
plt.grid(True)
plt.xlabel("Revenue")
plt.ylabel("departments")
plt.legend()
plt.show()
```
