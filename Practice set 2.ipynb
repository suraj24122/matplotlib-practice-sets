```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature = [14, 16, 20, 25, 30, 35, 33, 32, 28, 24, 18, 15]
```

```python
plt.plot(months, temperature, label = "Temperature Tracking", marker = "^", linestyle = "--", color = "red")
plt.title("Monthly average temperature tracking")
plt.xlabel("Months")
plt.ylabel("Temperature (C degree)")
plt.grid(True)
plt.legend()
plt.show()
```
